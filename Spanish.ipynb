```python
nlp = spacy.load("es_core_news_md")
nlp.add_pipe("textrank")
```

```python
nlp.max_length = 15000000
```

```python
def generate_final_summary(test_content, sum_mbart, sum_mt5):
  doc = nlp(test_content, disable = ['ner'])
  sentences = [sent.text for sent in doc.sents]
  summary_cont = lxr.get_summary(sentences, threshold=None)
  scores_cont = lxr.rank_sentences(sentences, threshold=None, fast_power_method=False,)
  ph_txt = []
  ph_rnk = []
  ph_di = {}
  for phrase in doc._.phrases:
    ph_txt.append(phrase.text)
    ph_rnk.append(phrase.rank)
    ph_di[phrase.text] = phrase.rank
  top_phs_1k = [x for _, x in sorted(zip(ph_rnk, ph_txt),  reverse=True)][:1000]
  new_scr = []
  for ss, lx_scr in zip(summary_cont, scores_cont):
    scr_tot = 0.0
    cnt = 0.0
    for ph in top_phs_1k:
      if ph in ss:
        cnt = cnt + 1.0
        scr_tot = scr_tot + ph_di[ph]
    if cnt>0:
      new_scr.append(lx_scr + (scr_tot/cnt))
    else:
      new_scr.append(lx_scr)
  sums_lex_text_rank_sorted = ' '.join([x for y, x in sorted(zip(new_scr, summary_cont),  reverse=True)])
  all_sum = sum_mbart + sum_mt5 + sums_lex_text_rank_sorted
  final_summ = ' '.join(all_sum.split(' ')[:1000])
  return  final_summ
```
